```python
import mysql.connector as msql
try:
    conn=msql.connect(host='localhost',database='cardataset',user='root',password='tamil')
    if conn.is_connected():
        cursor=conn.cursor()
        cursor.execute('select database()')
        record=cursor.fetchone()
        print("you're connected to the database",record)
        cursor.execute('drop table if exists car')
        print('creating a table')
        cursor.execute('create table car(buying varchar(30),maint varchar(30),doors varchar(30),persons varchar(30),lug_boot varchar(30),safety varchar(30),class varchar(30))')
        sql='insert into car values(%s,%s,%s,%s,%s,%s,%s)'
        with open(file=r'C:\Users\Tamilselvan.E\Downloads\car.data',mode='r') as f:
            for i in f.read().split(sep='\n'):
                data=tuple(i.split(sep=','))
                cursor.execute(sql,data)
                conn.commit()        
except Exception as e:
    print('Error:',e)
else:
    print('Data Loading process completed')
    f.close()
    cursor.close()
    conn.close()
```
