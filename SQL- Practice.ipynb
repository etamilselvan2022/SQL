```python
f=open (file=r'C:\Users\Tamilselvan.E\Downloads\car.data',mode='r')
sql='insert into car values (%s,%s,%s,%s,%s,%s,%s)'
try:
    for i in f.read().split(sep='\n'):
        l=i.split(sep=',')
        cursor.execute(sql,tuple(l))
        db.commit()
except Exception as e:
    print('Error',e)
f.close()    
```
